```python
import numpy as np
import matplotlib.pyplot as plt

from strategies import call_payoff, put_payoff, plot_strategy

S = np.linspace(0, 200, 500)
```

### Bull Call Spread

```python
K1, K2 = 100, 120
premium1, premium2 = 8, 3

long_call = call_payoff(S, K1, premium1, 'long')
short_call = call_payoff(S, K2, premium2, 'short')

plot_strategy(S, [long_call, short_call], ['Long Call (K=100)', 'Short Call (K=120)'], title="Bull Call Spread")
```

### Bear Put Spread

```python
K1, K2 = 120, 100
premium1, premium2 = 10, 4

long_put = put_payoff(S, K1, premium1, 'long')
short_put = put_payoff(S, K2, premium2, 'short')

plot_strategy(S, [long_put, short_put], ['Long Put (K=120)', 'Short Put (K=100)'], title="Bear Put Spread")
```

### Long Straddle

```python
K = 110
call_prem = 9
put_prem = 10

long_call = call_payoff(S, K, call_prem, 'long')
long_put = put_payoff(S, K, put_prem, 'long')

plot_strategy(S, [long_call, long_put], ['Long Call', 'Long Put'], title="Long Straddle")
```

### Long Strangle

```python
K_call = 120
K_put = 100
call_prem = 6
put_prem = 5

call_leg = call_payoff(S, K_call, call_prem, 'long')
put_leg = put_payoff(S, K_put, put_prem, 'long')

plot_strategy(S, [call_leg, put_leg], ['Long OTM Call', 'Long OTM Put'], title="Long Strangle")
```

### Butterfly Spread (Call)

```python
K1, K2, K3 = 100, 110, 120
premium1, premium2, premium3 = 5, 9, 3

long_call1 = call_payoff(S, K1, premium1, 'long')
short_call2 = call_payoff(S, K2, premium2, 'short') * 2
long_call3 = call_payoff(S, K3, premium3, 'long')

plot_strategy(S, [long_call1, short_call2, long_call3], 
              ['Long Call (K1)', 'Short 2x Call (K2)', 'Long Call (K3)'], 
              title="Call Butterfly Spread")
```

### Iron Condor

```python
K1, K2, K3, K4 = 90, 100, 120, 130
prem_put_sell, prem_put_buy = 5, 2
prem_call_sell, prem_call_buy = 4, 1

short_put = put_payoff(S, K2, prem_put_sell, 'short')
long_put = put_payoff(S, K1, prem_put_buy, 'long')
short_call = call_payoff(S, K3, prem_call_sell, 'short')
long_call = call_payoff(S, K4, prem_call_buy, 'long')

plot_strategy(S, [short_put, long_put, short_call, long_call],
              ['Short Put (K2)', 'Long Put (K1)', 'Short Call (K3)', 'Long Call (K4)'],
              title="Iron Condor")
```
