```python
import numpy as np
import matplotlib.pyplot as plt
from valuation_utils import (
    call_bounds, put_bounds,
    check_call_bounds, check_put_bounds,
    check_put_call_parity, plot_option_bounds
)
```

```python
S0 = 100      # Current stock price
K = 100       # Strike price
r = 0.05      # Risk-free interest rate
T = 1         # Time to maturity
C = 15        # Observed Call premium
P = 5         # Observed Put premium
```

```python
c_lower, c_upper = call_bounds(S0, K, r, T)
p_lower, p_upper = put_bounds(S0, K, r, T)

print(f"Call bounds: {c_lower:.2f} ≤ C ≤ {c_upper:.2f}")
print(f"Put bounds: {p_lower:.2f} ≤ P ≤ {p_upper:.2f}")
```

```python
call_ok, (c_l, c_u) = check_call_bounds(S0, K, r, T, C)
put_ok, (p_l, p_u) = check_put_bounds(S0, K, r, T, P)

print(f"Call Bound Check: {'OK' if call_ok else 'Violation'}")
print(f"Put Bound Check: {'OK' if put_ok else 'Violation'}")
```

```python
# ------------------------
# 3. Put-Call Parity Check
# ------------------------
parity_ok, (lhs, rhs) = check_put_call_parity(S0, K, r, T, C, P)
print(f"Put-Call Parity: {'OK' if parity_ok else 'Violation'}")
print(f"LHS (C - P) = {lhs:.2f}, RHS (S - K e^(-rT)) = {rhs:.2f}")
```

```python
S_vals = np.linspace(50, 150, 1000)
plot_option_bounds(S_vals, K, r, T, option_type="call")
plot_option_bounds(S_vals, K, r, T, option_type="put")
```
